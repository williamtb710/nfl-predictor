```python
import pandas as pd
import os

os.chdir("c:/Users/willi/nfl-predictor")

weekly = pd.read_parquet("data/raw/weekly_stats.parquet")

print(weekly.shape)
print(weekly.columns.tolist())
```

```python
#filter for wide receivers only
wr = weekly[weekly['position'] == 'WR'].copy()

print(f"WR rows: {len(wr):,}")
print(f"Unique WRs: {wr['player_name'].nunique()}")
print(f"\nSeasons covered: {sorted(wr['season'].unique())}")
print(f"Weeks range: {wr['week'].min()} to {wr['week'].max()}")

#frequency of zero receiving yard games
zero_pct = (wr['receiving_yards'] == 0).mean()
print(f"\nGames with zero receiving yards: {zero_pct:.1%}")

#basic stats for target variable
print(f"\nReceiving yards distribution:")
print(wr['receiving_yards'].describe().round(1))
```

```python
#missing-value check in relevant columns
key_cols = ['receiving_yards', 'targets', 'receptions', 'target_share', 'air_yards_share', 'receiving_air_yards', 'receiving_epa', 'wopr', 'opponent_team', 'recent_team', 'season', 'week']

missing = wr[key_cols].isnull().sum()
print("Missing values per column:")
print(missing)
```

```python
#sanity check for a single player across a season
player = wr[wr['player_name'] == 'J.Jefferson']
player = player[player['season'] == 2022].sort_values('week')
print(player[['season', 'week', 'recent_team', 'opponent_team', 
              'targets', 'receptions', 'receiving_yards', 
              'target_share', 'wopr']].to_string())
```

```python
missing_rows = wr[wr['target_share'].isnull()]
print(f"Total missing: {len(missing_rows)}")
print(f"\nSeasons affected:\n{missing_rows['season'].value_counts()}")
print(f"\nWeek numbers affected:\n{missing_rows['week'].value_counts().sort_index()}")
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Full distribution
axes[0].hist(wr['receiving_yards'], bins=50, edgecolor='black', color='steelblue')
axes[0].set_title('Receiving yards distribution (all games)')
axes[0].set_xlabel('Receiving yards')
axes[0].set_ylabel('Count')

# Exclude zero-yard games
nonzero = wr[wr['receiving_yards'] > 0]
axes[1].hist(nonzero['receiving_yards'], bins=50, edgecolor='black', color='steelblue')
axes[1].set_title('Receiving yards distribution (excluding zeros)')
axes[1].set_xlabel('Receiving yards')

plt.tight_layout()
plt.show()
```
